# 4.	Explain an adaptive thresholding-based binarization, with an example. Also, write a source code (in MATLAB/Python) 
for the same  the input image can be any (your choice).

Adaptive thresholding-based binarization is the method where the threshold value is calculated for smaller regions 
and therefore, there will be different threshold values for different regions. In OpenCV, you can perform adaptive threshold operations on an image using the method adaptive threshold() of the imgproc class


```python
# source code 
def questionFour(Original_image):
    gray_img = cv.imread(Original_image,cv.IMREAD_GRAYSCALE)
    
    fig,ax = plt.subplots(1,3)
    fig.set_figheight(5)
    fig.set_figwidth(20)
    
    blurred = cv.GaussianBlur(gray_img, (7, 7), 0)
    adaptive_thresh_gaussian = cv.adaptiveThreshold(blurred, 255,
	                  cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 119, 5)
    adaptive_thresh_mean = cv.adaptiveThreshold(blurred, 255,
	                  cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 119, 5)
        
    ax[0].axis('off')
    ax[0].imshow(gray_img,cmap='gray')
    ax[0].set_title('Original Image')

    ax[1].axis('off')
    ax[1].imshow(adaptive_thresh_gaussian,cmap='gray')
    ax[1].set_title('Adaptive Thresholding_Gaussian')
    
    ax[2].axis('off')
    ax[2].imshow(adaptive_thresh_mean,cmap='gray')
    ax[2].set_title('Adaptive Thresholding_Mean')
    

    plt.show()
    
questionFour("C:\\Users\\adede\\OneDrive\\Desktop\\sample2.jpg")
```

# 8. Explain correlation and convolution? Do you find any difference?

Convolution is the application of a dot product across an entire array with a kernel in order to create a new image based on the output of the dot products
output. For images there are typically either 1 or 3 channel colors with 2 dimensions one must take into consideration when doing convolution. For
simplicity 1 channel will be discussed here. To start out one could take a grayscale image and start with the first pixel and a kernel that is a 2d array
something like [[1,2,3],[1,2,3],[1,2,3]] and dot product that with the 8 pixels that surround the first pixel (in order to get 8 pixels around the first image we could
use a padding value of zero or something like the average of the pixels in the 9 pixels area) after the dot product is done shift over to the next pixel and do this
process for every pixel (this is one type of convolution). A single convolution will make a new array that is the same size as the old with new values. Convolutions
create filters for images and can be used to do things like blurring, sharpening, edge detection etc. The most important part of convolution is that it takes into
consideration the pixels around a given pixel allowing for the creation of new arrays that are based on this consideration.

The operation called correlation is closely related to convolution. In correlation, the value of an output pixel is also computed as a weighted sum of neighboring pixels. The difference is that the matrix of weights, in this case called the correlation kernel, is not rotated during the computation. The Image Processing Toolbox filter design functions return correlation kernels.

The basic different between the two is that Correlation is measurement of the similarity between two signals/sequences, while
Convolution is measurement of effect of one signal on the other signal.

# 9. Why is an image enhancement necessary? Explain it with an example (a source code may help)

The goal of image enhancement is to improve the usefulness of an image for a given task such as providing a more subjectively pleasing image for human viewing. In image enhancement, little or no attempt is made to estimate the actual image degradation process, and the techniques are often ad hoc. This process does not increase the inherent information content in data. It is a subjective process. It includes gray level and contrast manipulation, noise reduction, edge sharpening, filtering, interpolation and magnification, and pseudocoloring. Image enhancement techniques can be divided into two categories: frequency domain methods and spatial domain methods. The former process the image as a two-dimensional signal and enhance the image based on its two-dimensional Fourier transform. The low-pass filterbased method can remove noise from the image, whereas using high-pass filtering, we can enhance the edge, which is a kind of high-frequency signal, and make the blurred image clear.
